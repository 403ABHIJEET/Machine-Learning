```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

webpage = requests.get('https://www.scrapethissite.com/pages/simple/', headers=headers)
```

```python
webpage
```

```python
soup = BeautifulSoup(webpage.text, 'lxml')
```

```python
countries = []
for i in soup.find_all('h3'):
    countries.append(i.text.strip())
```

```python
population = []
for i in soup.find_all('span', class_ = 'country-population'):
    population.append(i.text.strip())
```

```python
for i in range(len(countries)):
    print("Country: " + countries[i])
    print("Population: " + population[i])
    print()
```
